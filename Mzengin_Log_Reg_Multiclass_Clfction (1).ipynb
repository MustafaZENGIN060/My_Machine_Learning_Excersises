```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv (r'C:\Users\ABC\Desktop\Yazılım\Machine Learning\Kendi Calismalarim\iris.csv')
```

```python
dataset['Species'].unique()
```

```python
dataset['Species'].replace({'Iris-setosa' : '1', 'Iris-versicolor' : '2', 'Iris-virginica': '3'},inplace = True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset[["Id", "SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]], dataset ["Species"], test_size=0.20)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model= LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
X_test
```

```python
model.score(X_test, y_test)
```
